```python
import pandas as pd
df=pd.read_csv("employee_dataset.csv")
df.head()
```

```python
#Checking Null values
df.isnull().sum()
```

```python
# Replacing null with mean of that column
df["PerformanceScore"].fillna(df["PerformanceScore"].mean(),inplace=True)
df["WorkHours"].fillna(df["WorkHours"].mean(),inplace=True)
#df["WorkHours"].fillna(0,inplace=True)
df["Salary"].fillna(df["Salary"].mean(),inplace=True)
df.isnull().sum()
```

```python
#checking for duplicate values
df.duplicated().sum()
```

```python
# Group by to calculate max salary dep wise 
print(df.groupby("Department")["Salary"].max())
```

```python
# Group by to calculate mean salary dep wise
print(df.groupby("Department")["Salary"].mean())
```

```python
# group by agg of salary and working hours
print(df.groupby("Department").agg({"Salary": ["mean", "max", "min", "count"], "WorkHours": ["mean", "max", "min", "count"]}))
```

```python
#histogram of salary
df["Salary"].plot(kind="hist", title="Salary Distribution", y="Frequency")
```

```python
#scatter plot of salary vs work hours
df.plot(kind="scatter", x="WorkHours", y="Salary", title="Salary vs Work Hours")
```

```python
# bar plot of average salary by department
df.groupby("Department")["Salary"].mean().plot(kind="bar", title="Average Salary by Department")
```
